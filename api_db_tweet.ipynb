```python
import pandas as pd
```

```python
import sqlite3

# Create / Connect Database

conn = sqlite3.connect('api_db_tweet.db') # Open a database File

print('Sukses')
```

```python
# CREATE TABLE text_tweet (input)

query = ''' create table text_tweet(
id INTEGER PRIMARY KEY AUTOINCREMENT,
tweet  varchar(255),
tweet_baru varchar(255)
)
 '''

conn.execute(query)

conn.commit()
```

```python
# CREATE TABLE upload_text_tweet (upload-post)

query = ''' create table upload_text_tweet(
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        tweet  varchar(255),
        tweet_baru varchar(255)
        )
        '''

conn.execute(query)

conn.commit()
```

```python
#CEK TABLE YG ADA DI DATABASE KITA
query = ''' SELECT name FROM sqlite_master WHERE type='table';
 '''

conn.execute(query)

conn.commit()
df = pd.read_sql_query(query, conn)
df.head(10)
```

```python
#BACA ISI DATA

query = ''' select * from text_tweet '''
df = pd.read_sql_query(query, conn)
df.head(50)
```

```python
#BACA ISI DATA

query = ''' select * from upload_text_tweet '''
df = pd.read_sql_query(query, conn)
df.head(50)
```
